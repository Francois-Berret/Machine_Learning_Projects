# Modèle de référence

```python
#Séparation des variables explicatives (X) et de la variable cible (y)

target = "Weekly_Sales"
"""
Imputer des dates présente un risque de biaiser les résultats de la prédiction, choix est fait de ne pas imputer et de ne garder que les
valeurs effectives au prix de la perte de 18 observations, ce qui ne change pas l'ordre de grandeur du dataset
"""
mask = ~(df["Year"].isna() & df["Month"].isna() & df["Day"].isna())
df_with_date = df.loc[mask,:]

X = df_with_date.drop(target, axis=1)
y = df_with_date.loc[:,target]

print("X :")
print(X.head(), "\n")
print("y :")
print(y.head(), "\n")
```

```python
#Création d'un train set et d'un test set au sein X et et de y

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Création des pipelines de preprocessing des variables quantitatives

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler


numeric_features = ["Temperature", "Fuel_Price", "CPI", "Unemployment", "Year", "Month", "Day"]
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
    ])
```

```python
#Création des pipelines de preprocessing des variables quantitatives

from sklearn.preprocessing import OneHotEncoder

categorical_features = ['Store', 'Holiday_Flag']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first'))
    ])
```

```python
#Instanciation du preprocessor à partir des pipelines suivant la nature des variables

from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
        ])
```

```python
#Application du preprocessing

print("Variables explicatives d'entraînement avant preprocessing : \n", X_train.head(), "\n")
X_train = preprocessor.fit_transform(X_train)
print("Variables explicatives d'entraînement après preprocessing : \n", X_train[0:5], "\n")

print("Variables explicatives de test avant preprocessing : \n", X_test.head(), "\n")
X_test = preprocessor.transform(X_test)
print("Variables explicatives de test après preprocessing : \n", X_test[0:5,:], "\n")
```

```python
#Entraînement du modèle sur les données d'entraînement

from sklearn.linear_model import LinearRegression

regressor1 = LinearRegression()
regressor1.fit(X_train, y_train)
```

```python
#Prédictions des valeurs cibles pour chaque observation d'entraînement

y_train_pred = regressor1.predict(X_train)
print(y_train_pred[:3])
```

```python
#Prédictions des valeurs cibles pour chaque observation de test

y_test_pred = regressor1.predict(X_test)
print(y_test_pred[:3])
```

```python
#Evalutation de la performance du modèle par R² sur les données d'entraînement et de test

from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor1, X_train, y_train, cv = 4, scoring = "r2")
print("R² de validation obtenus par validation croisée sur les données d'entraînement:", scores)
print("Moyenne : ", scores.mean().round(3))
print("Ecart-type : ", scores.std().round(3))
print("R² sur les données de test : ", regressor1.score(X_test, y_test).round(3))
```

```python
#Observation des coefficients des variables explicatives dans la prédiction des valeurs cibles

numeric_features_names = numeric_features
categorical_features_names = preprocessor.named_transformers_['cat']['encoder'].get_feature_names_out(categorical_features)
feature_names = numeric_features_names + list(categorical_features_names)

coefficients = regressor1.coef_

coefs_df = pd.DataFrame({'Variable': feature_names, 'Coefficient': coefficients.round(2)})
coefs_df
```

```python
#Visualisation de l'effet de chaque variable explicative sur la valeur prédite de la variable cible

fig = px.bar(y = coefs_df["Variable"], x = coefs_df["Coefficient"])
fig.update_layout(
    showlegend = True,
    margin = {'l': 120},
    title="Coefficients des variables explicatives",
    xaxis_title="Coefficient",
    yaxis_title="Variable")

fig.show()
```

# Modèle avec régularisation

```python
#GridSearch et validation croisée pour trouver le meilleur paramétrage du modèle régression linéaire régularisée Ridge

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

param_grid = {'alpha': [0.001, 0.01, 0.05, 0.1]}
regressor2 = Ridge()

grid_search1 = GridSearchCV(regressor2, param_grid, cv=4, scoring='r2')
grid_search1.fit(X_train, y_train)

print(grid_search1.best_estimator_)
print(grid_search1.best_score_.round(3))
print(grid_search1.cv_results_['std_test_score'].mean().round(3), "\n")
print("R² score on training set : ", grid_search1.score(X_train, y_train).round(3))
print("R² score on test set : ", grid_search1.score(X_test, y_test).round(3))
```

```python
#GridSearch et validation croisée pour trouver le meilleur paramétrage du modèle régression linéaire régularisée Lasso

from sklearn.linear_model import Lasso

param_grid = {'alpha': [500, 750, 1000]}
regressor3 = Lasso(max_iter=10000)

grid_search2 = GridSearchCV(regressor3, param_grid, cv=4, scoring='r2')
grid_search2.fit(X_train, y_train)

print(grid_search2.best_estimator_)
print(grid_search2.best_score_.round(3))
print(grid_search2.cv_results_['std_test_score'].mean().round(3), "\n")
print("R² score on training set : ", grid_search2.score(X_train, y_train).round(3))
print("R² score on test set : ", grid_search2.score(X_test, y_test).round(3))
```

```python
#Visualisation des R² scores de test et d'entraînement obtenus avec le modèle de base pour observer l'effet de la régularisation

print("R² sur les données d'entraînement' : ", regressor1.score(X_train, y_train).round(3))
print("R² sur les données de test : ", regressor1.score(X_test, y_test).round(3))
```
